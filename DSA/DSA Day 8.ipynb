```python
'''Question:
Given an array, implement the Bubble Sort algorithm and count the number of swaps made during the sorting process.

Input:
        * An array arr[] of integers of size n.

Output:
        * Return the number of swaps performed during the sorting process.

Constraints

        * 1≤n≤10^5
        * −10^6 ≤arr[i]≤10^6
Inputs:
arr = [64, 25, 12, 22, 11]
Output:
9'''

def bubble_sort_with_swap_count(arr):
    # WRITE YOUR CODE HERE...
    n = len(arr)
    swap_count = 0
    for i in range(n):
        swapped = False
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swap_count += 1
                swapped = True
        if not swapped:
            break
    return swap_count

arr = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]
print(bubble_sort_with_swap_count(arr))
```

```python
'''Question:
Implement the Bubble Sort algorithm to sort a given integer array in ascending order.

Input:
        * An array arr[] of integers of size n.

Output:
        * Return the sorted array in ascending order.

Constraints

        * 1≤n≤10^5
        * −10^6 ≤arr[i]≤10^6
Inputs:
arr = [64, 25, 12, 22, 11]
Output:
[11, 12, 22, 25, 64]
Explanation:
The ascending order of  [64, 25, 12, 22, 11] is [11, 12, 22, 25, 64]'''

def bubble_sort(arr):
    # WRITE YOUR CODE HERE...
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr

arr = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]
print(bubble_sort(arr))
```

```python
'''Question:
Implement the Insertion Sort algorithm and count the number of comparisons made during the sorting process.

Input:
        * An array arr[] of integers of size n.

Output:
        * Return the number of comparisons performed during the sorting process.

Constraints:
        * 1≤n≤10^5
        * −10^6 ≤arr[i]≤10^6
Inputs:
arr = [12, 11, 13, 5, 6]
Output:
7
Explanation:
the number of comparisons performed during the sorting process of  [12, 11, 13, 5, 6] is 7.'''

def insertion_sort_with_comparisons(arr):
    # WRITE YOUR CODE HERE...
    count=0
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            count+=1
            j -= 1
        arr[j + 1] = key
    return count

arr = [1, 2, 3, 4, 5]
print(insertion_sort_with_comparisons(arr))
```

```python
'''Question:
Implement the Insertion Sort algorithm to sort a given integer array in ascending order.

Input:
        * An array arr[] of integers of size n.

Output:
        * Return the sorted array in ascending order.

Constraints:
        * 1≤n≤10^5
        * −10^6 ≤arr[i]≤10^6
Inputs:
arr = [12, 11, 13, 5, 6]
Output:
[5, 6, 11, 12, 13]
Explanation:
Ascending order of [12, 11, 13, 5, 6] is [5, 6, 11, 12, 13]'''

def insertion_sort(arr):
    # WRITE YOUR CODE HERE...
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr
arr = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]
print(insertion_sort(arr))
```

```python
'''Question:
Implement the Selection Sort algorithm and find the Kth smallest element in the given array. The selection sort will sort the array partially, and once the K smallest elements are placed at the front, return the Kth element.

Input:
        * An array arr[] of integers of size n.
        * An integer k (1 ≤ k ≤ n), representing the index of the smallest element to find.

Output:
        * Return the kth smallest element after partially sorting the array.

Constraints:
        * 1≤n≤10^5
        * −10^6 ≤arr[i]≤10^6
Inputs:
arr = [12, 3, 5, 7, 19], k = 4
Output:
12'''

def kth_smallest_selection_sort(arr, k):
    # WRITE YOUR CODE HERE...
    n=len(arr)
    for i in range(n):
        min_index=i
        for j in range(i+1,n):
            if arr[j]<arr[min_index]:
                min_index=j
        arr[i],arr[min_index]=arr[min_index],arr[i]
    return arr[k-1]
arr = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]
k = 5
print(kth_smallest_selection_sort(arr, k))
```

```python
'''Question:
Given a sorted array, implement ternary search to check if an element x exists in the array. Return True if the element exists, otherwise return False.

Input:
        * A sorted integer array arr[] of size n.
        * An integer x representing the element to search for.

Output:
        * Return True if the element x exists in the array, otherwise return False.

Constraints:
        * 1≤n≤10^5
        * −10 6≤arr[i]≤10^6
        * −10 6≤x≤10^6
Inputs:
arr = [2, 4, 6, 8, 10], x = 6
Output:
True
Explanation:
The element 6 exists in the array so output is True'''

def ternary_search(arr, x):
    # WRITE YOUR CODE HERE...
    left,right=0,len(arr)-1
    while left<=right:
        mid1=left+(right-left)//3
        mid2=right-(right-left)//3
        if arr[mid1]==x:
            return True
        if arr[mid2]==x:
            return True
        if x<arr[mid1]:
            right=mid1-1
        elif x>arr[mid2]:
            left=mid2+1
        else:
            left = mid1 + 1
            right = mid2 - 1
    return False

def element_exists(arr, x):
    # WRITE YOUR CODE HERE...
    return ternary_search(arr, x)

arr = [1, 2, 3, 4, 5]
x = 2
print(element_exists(arr, x))
```

```python
'''Question:
Implement a ternary search algorithm to find the index of an element in a sorted array. If the element is not found, return -1.

Input:
        * A sorted integer array arr[] of size n.
        * An integer x representing the element to search for.

Output:
        * Return the index of the element x in the sorted array arr[] or -1 if the element is not found.

Constraints:
        * 1≤n≤10^5
        * −10 6≤arr[i]≤10^6
        * −10 6≤x≤10^6
Inputs:
arr = [1, 3, 5, 7, 9, 11], x = 7
Output:
3
Explanation:
the index of the element 7 in the sorted array  [1, 3, 5, 7, 9, 11] is 3'''

def ternary_search(arr, x):
    # WRITE YOUR CODE HERE...
    left,right=0,len(arr)-1
    while left<=right:
        mid1=left+(right-left)//3
        mid2=right-(right-left)//3
        if arr[mid1]==x:
            return mid1
        if arr[mid2]==x:
            return mid2
        if x<arr[mid1]:
            right=mid1-1
        elif x>arr[mid2]:
            left=mid2+1
        else:
            left = mid1 + 1
            right = mid2 - 1
    return -1
arr = [1, 2, 3, 4, 5]
x = 4
print(ternary_search(arr, x))
```

```python
'''Question:
Implement the Selection Sort algorithm to sort a given array of integers in ascending order. This algorithm works by repeatedly finding the minimum element from the unsorted part and swapping it with the leftmost unsorted element.

Input:
        * An array arr[] of integers of size n.

Output:
        * Return the sorted array in ascending order.

Constraints:
        * 1≤n≤10^5
        * −10^6 ≤arr[i]≤10^6
Inputs:
arr = [64, 34, 25, 12, 22, 11, 90]
Output:
[11, 12, 22, 25, 34, 64, 90]
Explanation:
The Ascending order of [64, 34, 25, 12, 22, 11, 90] is [11, 12, 22, 25, 34, 64, 90]'''

def selection_sort(arr):
    # WRITE YOUR CODE HERE...
    n=len(arr)
    for i in range(n):
        min_index=i
        for j in range(i+1,n):
            if arr[j]<arr[min_index]:
                min_index=j
        arr[i],arr[min_index]=arr[min_index],arr[i]
    return arr
arr = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]
print(selection_sort(arr))
```

```python
'''Question:
Implement the Merge Sort algorithm and count the number of inversions in the array. An inversion is a pair (i, j) such that i < j and arr[i] > arr[j].

Input:
        * An array arr[] of integers of size n.

Output:
        * Return the total number of inversions in the array.

Constraints:
        * 1≤n≤10^5
        * −10^6 ≤arr[i]≤10^6
Inputs:
arr = [2, 3, 8, 6, 1], left = 0, right = 4
Output:
3
Explanation:
The total number of inversions in the array is 3'''

def merge(arr, left, mid, right):
    left_sub = arr[left:mid+1]
    right_sub = arr[mid+1:right+1]

    i = 0
    j = 0
    k = left
    inv_count = 0

    while i < len(left_sub) and j < len(right_sub):
        if left_sub[i] <= right_sub[j]:
            arr[k] = left_sub[i]
            i += 1
        else:
            arr[k] = right_sub[j]
            inv_count += (len(left_sub) - i)
            j += 1
        k += 1

    while i < len(left_sub):
        arr[k] = left_sub[i]
        i += 1
        k += 1

    while j < len(right_sub):
        arr[k] = right_sub[j]
        j += 1
        k += 1

    return inv_count

def merge_and_count(arr, left, right):
    inv_count = 0
    if left < right:
        mid = (left + right) // 2
        inv_count += merge_and_count(arr, left, mid)
        inv_count += merge_and_count(arr, mid + 1, right)
        inv_count += merge(arr, left, mid, right)
    return inv_count

arr = [1, 1, 1, 1]
left = 1
right = 1
print(merge_and_count(arr, left, right))
```

```python
'''Question:
Implement the Merge Sort algorithm to sort a given array of integers in ascending order. Merge Sort is a divide and conquer algorithm that splits the array in half, recursively sorts the halves, and then merges them back together.

Input:
        * An array arr[] of integers of size n.

Output:
        * Return the sorted array in ascending order.

Constraints:
        * 1≤n≤10^5
        * −10^6 ≤arr[i]≤10^6
Inputs:
arr = [38, 27, 43, 3, 9, 82, 10]
Output:
[3, 9, 10, 27, 38, 43, 82]
Explanation:
The Ascending order of [38, 27, 43, 3, 9, 82, 10] is [3, 9, 10, 27, 38, 43, 82]'''

def merge_sort(arr):
    # WRITE YOUR CODE HERE...
    if len(arr)<=1:
        return arr
    mid=len(arr)//2
    left_half=merge_sort(arr[:mid])
    right_half=merge_sort(arr[mid:])
    return merge(left_half,right_half)

def merge(left,right):
    result=[]
    i=j=0
    while i<len(left) and j<len(right):
        if left[i]<=right[j]:
            result.append(left[i])
            i+=1
        else:
            result.append(right[j])
            j+=1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

arr = [1, -1, 2, -2, 3, -3]
print(merge_sort(arr))
```
