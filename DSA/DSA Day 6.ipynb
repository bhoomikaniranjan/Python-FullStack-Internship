```python
'''Question:
Design a queue data structure using a list with the following operations:
        * enqueue(x) - Add an element x to the back of the queue.
        * dequeue() - Remove and return the front element of the queue. If the queue is empty, return -1.

Input:
        * Sequence of operations (enqueue or dequeue).

Output:
        * The result of dequeue operations.

Constraints:
        * Number of operations 1≤N≤10 5
        * Elements in enqueue −10 6 ≤x≤10 6
        * Operations are valid.
Inputs:
enqueue(1),  enqueue(2), dequeue(),  enqueue(3),  dequeue(),  dequeue()
Output:
1, 2, 3'''

class Queue:
    def __init__(self):
        # WRITE YOUR CODE HERE...
        self.in_stack = []
        self.out_stack = []

    def enqueue(self, x):
        # WRITE YOUR CODE HERE...
        self.in_stack.append(x)

    def dequeue(self):
        # WRITE YOUR CODE HERE...
        if not self.out_stack:
            while self.in_stack:
                self.out_stack.append(self.in_stack.pop())
        return self.out_stack.pop() if self.out_stack else -1

q = Queue()
q.enqueue(100)
q.enqueue(200)
print(q.dequeue())
q.enqueue(300)
print(q.dequeue())
print(q.dequeue())
```

```python
'''Question:
Check if a queue is empty after performing a series of operations.

Input:
        * Sequence of operations (enqueue or dequeue).

Output:
        * Return True if the queue is empty, otherwise False.

Constraints:
        * Number of operations 1≤N≤10 5
        *  Operations are valid.
Inputs:
q.enqueue(1), q.dequeue()
Output:
True'''

class Queue:
    def __init__(self):
        # WRITE YOUR CODE HERE...
        self.in_stack = []
        self.out_stack = []

    def enqueue(self, x):
        # WRITE YOUR CODE HERE...
        self.in_stack.append(x)

    def dequeue(self):
        # WRITE YOUR CODE HERE...
        if not self.out_stack:
            while self.in_stack:
                self.out_stack.append(self.in_stack.pop())
        if self.out_stack:
            self.out_stack.pop()

    def is_empty(self):
        # WRITE YOUR CODE HERE...
        return not self.in_stack and not self.out_stack

q = Queue()
q.enqueue(10)
q.enqueue(20)
q.enqueue(10)
q.dequeue()
print(q.is_empty())
```

```python
'''Question:
Given two sorted singly linked lists, merge them into a single sorted linked list.

Input:
        * Two sorted linked lists with 1 ≤ number of nodes ≤ 10⁴.
        * Each node's value is between -10⁴ and 10⁴.

Output:
        * Return the head of the merged linked list.

Constraints:
        * Each linked list contains 1 ≤ number of nodes ≤ 10⁴.
        * Node values are between -10⁴ and 10⁴.
        * The merged list should also be sorted in ascending order.
        * The solution must run in O(n+m) time, where n and m are the sizes of the two lists.
Inputs:
l1 = create_linked_list([1, 3, 5, 7]), l2 = create_linked_list([2, 4, 6, 8])
Output:
1 -> 2 -> 3 -> 4 -> 5 -> 6 -> 7 -> 8
Explanation:
Merged Linked list of [1, 3, 5, 7] and [2, 4, 6, 8] is 1 -> 2 -> 3 -> 4 -> 5 -> 6 -> 7 -> 8'''

class ListNode:
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next

def merge_two_sorted_lists(l1, l2):
    # WRITE YOUR CODE HERE...
    dummy = ListNode()
    current = dummy

    while l1 and l2:
        if l1.value < l2.value:
            current.next = l1
            l1 = l1.next
        else:
            current.next = l2
            l2 = l2.next
        current = current.next
    if l1:
        current.next = l1
    elif l2:
        current.next = l2

    return dummy.next

def create_linked_list(values):
    dummy = ListNode()
    current = dummy
    for value in values:
        current.next = ListNode(value)
        current = current.next
    return dummy.next

def print_linked_list(head):
    current = head
    while current:
        print(current.value, end=" -> " if current.next else "\n")
        current = current.next

l1 = create_linked_list([])
l2 = create_linked_list([2, 4, 6])
merged_list = merge_two_sorted_lists(l1, l2)
print_linked_list(merged_list)

```

```python
'''Question:
Given a singly linked list, reverse it and return the new head of the reversed linked list.

Input:
        * A linked list where each node contains an integer value.
        * The number of nodes in the list is between 1 and 10⁴.
        * Each node's value is between -10⁴ and 10⁴.

Output:
        * Return the head of the reversed linked list.

Constraints:
        * The linked list contains 1 ≤ number of nodes ≤ 10⁴.
        * Each node's value is between -10⁴ and 10⁴.
        * The solution should run in O(n) time with O(1) space (in-place reversal).
Inputs:
head = ListNode(1, ListNode(2, ListNode(3, ListNode(4, ListNode(5))))), new_head = reverse_linked_list(head)
Output:
5 -> 4 -> 3 -> 2 -> 1
Explanation:
The linked list is reversed.'''

class ListNode:
    def __init__(self, value=0, next=None):
        # WRITE YOUR CODE HERE...
        self.value = value
        self.next = next

def reverse_linked_list(head):
    # WRITE YOUR CODE HERE...
    prev = None
    current = head
    while current:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
    return prev

def print_linked_list(head):
    # WRITE YOUR CODE HERE...
    while head:
        print(head.value, end=' -> ' if head.next else '\n')
        head = head.next

head = ListNode(1)
new_head = reverse_linked_list(head)
print(print_linked_list(new_head))
```

```python
'''Question:
Given a singly linked list, determine if there is a cycle. A cycle exists if any node is visited more than once while traversing.

Input:
        * A singly linked list with 1 ≤ number of nodes ≤ 10⁴.
        * Each node's value is between -10⁴ and 10⁴.

Output:
        * Return True if a cycle exists, otherwise False.

Constraints:

        * The linked list contains 1 ≤ number of nodes ≤ 10⁴.
        * Each node's value is between -10⁴ and 10⁴.
        * The solution must run in O(n) time with O(1) space (using two pointers).
Inputs:
head = ListNode(1, ListNode(2, ListNode(3, ListNode(4)))), head.next.next.next.next = head.next
Output:
True
Explanation:
The linked list contains a cycle.'''

class ListNode:
    def __init__(self, value=0, next=None):
        # WRITE YOUR CODE HERE...
        self.value = value
        self.next = next

def detect_cycle(head):
    # WRITE YOUR CODE HERE...
    if not head:
        return False
    slow = head
    fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True

    return False

head = ListNode(1, ListNode(2))
print(detect_cycle(head))
```

```python
'''Question:
Given a doubly linked list and a position, insert a new node with a given value at that position.

Input:
        * Head of the doubly linked list.
        * An integer value representing the value of the new node.
        * An integer position where the new node should be inserted.
        * 1≤length of DLL≤10^4
        * 1≤position≤length of DLL+1
        * −10^4≤value≤10^4

Output:
        * The updated head of the doubly linked list.

Constraints:
        * The position is guaranteed to be valid.
        * Maintain the structure of the doubly linked list.
Inputs:
head = create_dll([1, 2, 4]), value = 3, position = 3
Output:
[1, 2, 3, 4]
Explanation:
The updated head of the[1, 2, 4] doubly linked list is [1, 2, 3, 4]
'''
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
        self.prev = None

def insert_at_position(head, value, position):
    new_node = Node(value)
    if position == 1:
        new_node.next = head
        if head:
            head.prev = new_node
        head = new_node
        return head

    current = head
    count = 1
    while current and count < position - 1:
        current = current.next
        count += 1
    if current:
        new_node.next = current.next
        if current.next:
            current.next.prev = new_node
        new_node.prev = current
        current.next = new_node
    return head

def create_dll(arr):
    if not arr:
        return None
    head = Node(arr[0])
    current = head
    for value in arr[1:]:
        new_node = Node(value)
        current.next = new_node
        new_node.prev = current
        current = new_node
    return head

def display_dll(head):
    result = []
    while head:
        result.append(head.value)
        head = head.next
    return result

head = create_dll([])
value = 1
position = 1
print(display_dll(insert_at_position(head, value, position)))
```

```python
'''Question:
Given a circular linked list, insert a new node with a given value at a specific position.

Input:
        * The head of the circular linked list.
        * An integer value representing the value of the new node.
        * An integer position where the new node should be inserted.
        * 1≤length of CLL≤10^4
        * 1≤position≤length of CLL+1
        * -10^4≤value≤10^4

Output:
        * The updated head of the circular linked list.

Constraints:
        * Position is guaranteed to be valid.
        * Maintain the circular structure of the linked list.
Inputs:
cll = [1, 2, 4], value = 3, position = 3
Output:
[1, 2, 3, 4]
Explanation:
The updated head of the circular linked list is [1, 2, 3, 4].'''

class Node:
    def __init__(self, value):
        # WRITE YOUR CODE HERE...
        self.value = value
        self.next = None

def insert_at_position(head, value, position):
    # WRITE YOUR CODE HERE...
    new_node = Node(value)
    if not head:
        new_node.next = new_node
        return new_node
    if position == 1:
        current = head
        while current.next != head:
            current = current.next
        current.next = new_node
        new_node.next = head
        return new_node
    current = head
    for _ in range(position - 2):
        current = current.next
    new_node.next = current.next
    current.next = new_node
    return head

def create_cll(arr):
    if not arr:
        return None
    head = Node(arr[0])
    current = head
    for value in arr[1:]:
        new_node = Node(value)
        current.next = new_node
        current = new_node
    current.next = head
    return head

def display_cll(head):
    if not head:
        return []
    result = []
    current = head
    while True:
        result.append(current.value)
        current = current.next
        if current == head:
            break
    return result

cll = [1, 2, 4]
value = 6
position = 5
head = create_cll(cll)
print(display_cll(insert_at_position(head, value, position)))
```

```python
'''Question:
Delete the first occurrence of a given value from a doubly linked list.

Input:
        * Head of the doubly linked list.
        * An integer value to delete.
        * 1≤length of DLL≤ 10^4
        * −10^4 ≤value≤ 10^4

Output:
        * The updated head of the doubly linked list.

Constraints:
        * If the value does not exist, return the original list.
        * Maintain the structure of the doubly linked list.
Inputs:
list1 = [1, 2, 3, 4], value_to_delete = 3
Output:
[1, 2, 4]'''

class Node:
    def __init__(self, value):
        # WRITE YOUR CODE HERE...
        self.value = value
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        # WRITE YOUR CODE HERE...
        self.head = None

    def append(self, value):
        # WRITE YOUR CODE HERE...
        new_node = Node(value)
        if not self.head:
            self.head = new_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node
        new_node.prev = current


    def to_list(self):
        # WRITE YOUR CODE HERE...
        result = []
        current = self.head
        while current:
            result.append(current.value)
            current = current.next
        return result

def delete_by_value(head, value):
    # WRITE YOUR CODE HERE...
    current = head
    while current:
        if current.value == value:
            if current.prev:
                current.prev.next = current.next
            else:
                head = current.next
            if current.next:
                current.next.prev = current.prev
            break
        current = current.next
    return head

list1 = [1, 2, 3, 4]
dll = DoublyLinkedList()
for val in list1:
    dll.append(val)

value_to_delete = 4
dll.head = delete_by_value(dll.head, value_to_delete)
print(dll.to_list())
```

```python
'''Question:
You are given a set of jobs, each with a start time, end time, and a profit. Your goal is to find the maximum profit by selecting non-overlapping jobs.

Input:
        * A list jobs of tuples where each tuple (start_time, end_time, profit) represents a job with a start time, end time, and profit.
        * 1≤len(jobs)≤10^4
        * 1≤start time,end time,profit≤10^4

Output:
        * The maximum profit from selecting non-overlapping jobs.

Constraints:
        * Jobs should be selected such that no two jobs overlap.
Inputs:
jobs = [(1, 4, 50), (3, 5, 20), (6, 7, 70), (2, 6, 30), (5, 8, 60)]
Output:
120
Explanation:
The total profit is 50 + 70 = 120'''

import bisect
def job_scheduling(jobs):
    # WRITE YOUR CODE HERE...
    jobs.sort(key=lambda x: x[1])
    start_times = [job[0] for job in jobs]
    dp = [(0, 0)]
    for start, end, profit in jobs:
        i = bisect.bisect_right(dp, (start, float('inf'))) - 1
        current_profit = dp[i][1] + profit
        if current_profit > dp[-1][1]:
            dp.append((end, current_profit))
    return dp[-1][1]

jobs = [(1, 3, 20), (2, 4, 25), (4, 6, 10), (6, 8, 30)]
print(job_scheduling(jobs))
```

```python
'''Question:
You are given a list of jobs with deadlines and profits. You can schedule multiple jobs on the same day. Your task is to maximize the total profit by selecting jobs within their respective deadlines.

Input:
        * A list jobs of tuples where each tuple (profit, deadline) represents a job with a certain profit and deadline.
        * 1≤len(jobs)≤10^4
        * 1≤profit,deadline≤10^4

Output:
        * The maximum profit that can be earned by scheduling jobs within the deadlines.

Constraints:
        * You can schedule multiple jobs on the same day, but each job must be completed before its respective deadline.
Inputs:
jobs = [(100, 2), (200, 1), (150, 2), (300, 1)]
Output:
450
Explanation:
Job 1 with profit 200 and job 2 with profit 300 can be scheduled.'''

def job_scheduling(jobs):
    # WRITE YOUR CODE HERE...
    jobs.sort(reverse=True)
    total_profit = 0
    jobs_done = 0
    for profit, deadline in jobs:
        if deadline > jobs_done:
            total_profit += profit
            jobs_done += 1
    return total_profit

jobs = [(10, 3), (10, 1), (10, 2), (10, 4)]
print(job_scheduling(jobs))
```
